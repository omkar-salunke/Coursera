```
import numpy as np
import matplotlib.pyplot as plt

# Data
n = 10  # Total tosses
k = 7   # Heads observed

# Likelihood function
def likelihood(p):
    return (p**k) * ((1 - p)**(n - k))

# Values of p to evaluate
p_values = np.linspace(0, 1, 100)
likelihood_values = [likelihood(p) for p in p_values]

# Plotting the likelihood function
plt.figure(figsize=(8, 5))
plt.plot(p_values, likelihood_values, label="Likelihood", color="blue")
plt.axvline(0.7, color="red", linestyle="--", label="MLE: p = 0.7")
plt.title("Likelihood Function for Coin Bias")
plt.xlabel("Probability of Heads (p)")
plt.ylabel("Likelihood")
plt.legend()
plt.show()
```
